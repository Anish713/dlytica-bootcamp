```python
# import visualization libraries
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
```

```python
sorted_df=pd.read_csv('CleanedMovies.csv')
sorted_df.head()
```

```python
##plot the graph between year and gross_in_millions
#using line chart with dots
movies_after_2015 = sorted_df[sorted_df["movie_release_year"] >= 2015]
movies_after_2015
```

```python
sorted_df.columns
```

```python
df_TotalGross = movies_after_2015.groupby("movie_release_year")["gross_in_millions($)"].agg(sum).reset_index()
df_TotalGross
plt.plot(df_TotalGross["movie_release_year"], df_TotalGross["gross_in_millions($)"], 'o-')
plt.xlabel('Movie Release Year')
plt.ylabel('Total Gross Amount in Millions($)')
plt.title("Total gross in millions for movies released after 2015")
plt.show()
```

```python
# top 10 most rated movies
top_10 = sorted_df.sort_values(by='movie_rating', ascending=False).head(10)
# creating a bar chart for top 10 movies with highest rating
plt.barh(top_10["movie_title"], top_10["movie_rating"],color='r')
# plt.xticks(rotation=90)
plt.xlabel('Movie Rating')
plt.ylabel('Movie Title')
plt.title('Top 10 Movies with Highest Ratings')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Heatmap showing correlation among the numeric columns
# Assuming "sorted_df" is your DataFrame
numeric_columns = sorted_df.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()

# Create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, cmap='Greens', annot=False)

plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# checking for outliers
cols = ['movie_rating', 'gross_in_millions($)',  'runtime_in_minutes']
plt.figure(figsize=(20, 5))
plt.title("Showing outliers using Box Plot")
sns.boxplot(data=sorted_df[cols], orient='h')
```
